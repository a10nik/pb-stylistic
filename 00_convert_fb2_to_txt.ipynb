```python
import zipfile
import os
import lxml.etree as ET
```

```python
directory = './fb2'
for b in os.listdir(directory):
#     print(b)
    if not b.endswith('.fb2.zip'):
        continue
    
    zip_ref = zipfile.ZipFile(os.path.join(directory, b), 'r')
    zip_ref.extractall('fb2.unzipped')
    zip_ref.close()
```

```python
directory = './fb2.unzipped'
for b in os.listdir(directory):
    book_id = ".".join(b.rsplit('.', 2)[-2:])
    os.rename(os.path.join(directory, b), os.path.join(directory, book_id))
    
```

```python
if not os.path.exists("txt"):
    os.makedirs("txt")

fb2_to_txt_xslt = ET.parse("FB2_2_txt.xsl")

for xml_filename in os.listdir(directory):
    print(xml_filename)
    dom = ET.parse(os.path.join(directory, xml_filename))
    transform = ET.XSLT(fb2_to_txt_xslt)
    txt = transform(dom)
    name = os.path.basename(xml_filename)
    with open("txt/{}.txt".format(name), "w", encoding="utf-8") as txt_file:
        txt_file.write(str(txt))
```
