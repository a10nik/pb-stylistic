```python
dir_in = 'txt_cut'

char_range = []

for filename in sorted(os.listdir(dir_in)):
    with open(os.path.join(dir_in, filename), 'r') as book:
        size = len(book.read())
        if size <10:
            print(filename)
        char_range.append(size)
        
char_range = pd.Series(char_range)
```

```python
char_range.hist(bins=20)
```

```python
char_range.describe()
```
