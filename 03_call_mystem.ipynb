```python
import subprocess
import os
import time
```

```python
dir_in = 'txt_cut'
dir_out = 'json_cut'
```

```python
%%time
processes = set()
max_processes = 5

if not os.path.exists(dir_out):
    os.makedirs(dir_out)
n=0
for f in os.listdir(dir_in):
    if (f[-3:] == "txt"):
        json_file_name = os.path.join(dir_out, f + ".json")
        if not os.path.isfile(json_file_name):
            processes.add(subprocess.Popen(["./mystem", "--format=json", "-nigfcsd", os.path.join(dir_in, f), json_file_name]))
            while len(processes) >= max_processes:
                time.sleep(0.2)
                processes.difference_update(
                    [p for p in processes if p.poll() is not None])
            n+=1
            print(n, "creating", json_file_name)
        else:
            print("exists", json_file_name)
    
#Check if all the child processes were closed
for p in processes:
    if p.poll() is None:
        p.wait()
```
